```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def noise(t):
    return np.random.rand(len(t))

def square(t,amp,per,duty):
    return amp*np.ceil((t/per)%1+duty-1)

t = np.arange(0,10,0.01)
sig1 = square(t,1,2,0.2)+0.05*noise(t)
sig2 = square(t-0.23,1.4,2,0.2)+0.03*noise(t)

plt.subplot(2,1,1)
plt.plot(t,sig1)
plt.subplot(2,1,2)
plt.plot(t,sig2)
plt.show()
```
