```python
'''
Q.1 천체의 궤도 1

아래 코드는 태양을 원점으로 하여 어떤 천체의 궤도를 그린 것입니다.
그림에서 붉은색 점이 태양, 파란색 점이 지구이고 검은색 선이 천체의 궤도입니다.

현재 지구의 위치가 (sin(π/3),cos(π/3))의 위치에 있다고 하죠.
만약 천체가 아주 아주 빨라서 지구가 상대적으로 멈춰있다고 치면, 이 천체가 가장 지구에 근접한 거리는 얼마일까요?
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt

y = np.linspace(-3,3,200)
x = (y**2)/4-1

plt.plot(x,y,'k')
plt.plot(np.cos(np.pi/3),np.sin(np.pi/3),'.b')
plt.plot(0,0,'.r')
plt.show()
```

```python
#여기에 코드 입력
```

```python
'''
Q.2 천체의 궤도 2

똑같은 문제입니다. 하지만 이번엔 천체가 충분히 느리다고 해보죠.
천체의 궤도를 그릴때 리스트에 200개의 성분이 있었죠?
그 200개의 리스트가 지날 동안 지구는 한번에 반시계 방향으로 각도가 π/200씩 움직인다고 해봅시다.
(지구는 반지름이 1인 원운동을 한다고 합시다.)
그러면 아래의 그림처럼 궤도가 그려질 것입니다.

이런 상황에서 천체가 지구에 가장 근접한 거리는 얼마일까요?
'''
```

```python
t = [np.pi/3 + i*np.pi/200 for i in range(201)]

plt.plot(x,y,'k')
plt.plot(np.cos(t),np.sin(t),'b')
plt.plot(0,0,'.r')
plt.show()
```

```python
'''
Q.3 신호 처리

아래의 신호는 저항과 캐패시터 등으로 조금 '조작된' 사각파 신호입니다.
아마 이것이 실제 신호와 비슷한 모습일 것인데, 이것의 주기를 한번 구해봅시다.
'''
```

```python
import numpy as np
import matplotlib.pyplot as pyplot

def square(t,per,amp,duty):
    return amp*np.ceil((t/per)%1+duty-1)

def noize(t,gain):
    return gain*(np.random.rand(len(t)))

def decay(t,per,tau):
    return np.exp(-(t%per)/tau)

def attack(t,per,amp,duty,dt):
    r = (1-square(t+per*(1-duty),per,amp,duty))*(1-square(t+dt,per,amp,duty))
    return (amp-(np.expm1((-t)%per/dt))/(np.exp(1)-1))*r%amp

tick = np.arange(0,10,0.01)
sig = (1-square(tick-0.2,2.4,1,0.7))*decay(tick-0.2,2.4,0.25)+attack(tick-0.2,2.4,1,0.7,0.5)+noize(tick,0.07)

plt.plot(tick,sig)
plt.show()
```
